```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd
```

```python
my_service = Service(executable_path='c:/chrome_driver/chromedriver.exe')
browser = Browser('chrome', service=my_service)
```

```python
url = 'https://www.martincitybrewingcompany.com/beers'
browser.visit(url)
```

```python
beer_soup = soup(browser.html)
```

```python
for i in beer_soup.find_all('div',class_='sqs-html-content'):
    try:
        info = (i.find('p',style="text-align:center;white-space:pre-wrap;").text)
    except:
        pass
```

```python
info = info.split('    •  ')
address = info[0]
contact = info[1]
```

```python
brewery_dict = {}

brewery_dict['Name'] = 'Martin City Brewing Company'
brewery_dict['Address'] = address
brewery_dict['Contact'] = contact
brewery_dict['BreweryID'] = 4
martin_city_df = pd.DataFrame([brewery_dict])
martin_city_df
```

```python
martin_city_df.to_csv('MartinCityInfo.csv')
```

```python
beer_links = []
brews = []
for beer in beer_soup.find_all('section'):
    for link in beer.find_all('a',class_='gallery-grid-image-link'):
        brew = link['href']
        brews.append(brew)
        beer_links.append(f'https://www.martincitybrewingcompany.com{brew}')
```

```python
beer_links
```

```python
beer_list = []
for beer in beer_links:
    beer_dict = {}
    browser.visit(beer)
    time.sleep(1)
    beer_soup = soup(browser.html)
    beer_type = beer_soup.find('h3').text
    beer_dict["Name"] = beer_type
    for item in beer_soup.find_all('p',style="text-align:center;white-space:pre-wrap;"):
        if '%' in item.text:
            abv = item.text.replace('%','')
            beer_dict['Type'] = ''
            beer_dict['ABV'] = abv
    beer_list.append(beer_dict)
    
beer_list
```

```python
df = pd.DataFrame(beer_list)
df
```

```python
names = []
for full_name in df.iterrows():
    names.append(full_name[1]['Name'])
```

```python
style_parts = []
for name in names:
    words=name.split()
#     print(words)
    l = (len(words))
    style_part = []
    if l == 1:
        style_part.append(name)
        style_parts.append(style_part)
    if l == 2:
        style_part.append(name)
        for i in range(0,l):
            style_part.append(words[i])
        style_parts.append(style_part)
        
    if l == 3:
        style_part.append(name)
        for i in range(0,l):
            style_part.append(words[i])
        for i in range(0,l-1):
            style_part.append(words[i]+' '+words[i+1])
        style_parts.append(style_part)
    if l == 4:
        style_part.append(name)
        for i in range(0,l):
            style_part.append(words[i])
        for i in range(0,l-1):
            style_part.append(words[i]+' '+words[i+1])
        for i in range(0,l-2):
            style_part.append(words[i]+' '+words[i+1]+' '+words[i+2])
        style_parts.append(style_part)
#     if l == 5:
#         style_part.append(name)
        for i in range(0,l):
            style_part.append(words[i])
        for i in range(0,l-1):
            style_part.append(words[i]+' '+words[i+1])
        for i in range(0,l-2):
            style_part.append(words[i]+' '+words[i+1]+' '+words[i+2])
        for i in range(0,l-3):
            style_part.append(words[i]+' '+words[i+1]+' '+words[i+2]+' '+words[i+3])
        style_parts.append(style_part)

print(style_parts)
```

```python
styles = []
```

```python
bbeer_df = pd.read_csv('BoulevardBeers.csv',index_col=None)
bbeer_df = bbeer_df.drop(columns=['Unnamed: 0'])
bbeer_df = bbeer_df.rename(columns={"Type": "Style"})
for style in bbeer_df.iterrows():
    styles.append(style[1]['Style'])
```

```python
beer_styles = []
for style in style_parts:
    for item in style:
        for _item_ in styles:
            if item in _item_:
                beer_styles.append(item)
```

```python
beer_styles = list(set(beer_styles))
```

```python
temp_list = ['Apple','Winter','Hard','Warmer','Winter Warmer','Pale','Belgian','Hop','Style','Coffee','Hazy','Barrel','with']
```

```python
for style in temp_list:
    beer_styles.remove(style)
    
```
