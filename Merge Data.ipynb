```python
import pandas as pd
```

```python
bbeer_df = pd.read_csv('BoulevardBeers.csv',index_col=None)
bbeer_df = bbeer_df.drop(columns=['Unnamed: 0'])
bbeer_df = bbeer_df.rename(columns={"Type": "Style"})
bbeer_df['BreweryID'] = [1 for i in range(0,len(list(bbeer_df['Name'])))]
bbeer_df
```

```python
cbbeer_df = pd.read_csv('CityBarrelBeers.csv',index_col=None)
cbbeer_df = cbbeer_df.drop(columns=['Unnamed: 0'])
cbbeer_df = cbbeer_df.rename(columns={"Type": "Style"})
cbbeer_df['BreweryID'] = [2 for i in range(0,len(list(cbbeer_df['Name'])))]
cbbeer_df
```

```python
kcbeer_df = pd.read_csv('KcBierCoBeers.csv',index_col=None)
kcbeer_df = kcbeer_df.drop(columns=['Unnamed: 0'])
kcbeer_df = kcbeer_df.rename(columns={"Type": "Style"})
kcbeer_df['BreweryID'] = [3 for i in range(0,len(list(kcbeer_df['Name'])))]
kcbeer_df
```

```python
mcbeer_df = pd.read_csv('MarinCityBeers.csv',index_col = None)
mcbeer_df = mcbeer_df.drop(columns=['Unnamed: 0'])
mcbeer_df = mcbeer_df.rename(columns={"Type": "Style"})
mcbeer_df['BreweryID'] = [4 for i in range(0,len(list(mcbeer_df['Name'])))]
mcbeer_df
```

```python
beer_df = bbeer_df.append(cbbeer_df, ignore_index=True)
beer_df = beer_df.append(kcbeer_df,ignore_index=True)
beer_df = beer_df.append(mcbeer_df,ignore_index=True)
beer_df
```

```python
beer_df['BeerID'] = [i for i in range(0,len(list(beer_df['Name'])))]
beer_df
```

```python
beer_df.to_csv('Beers.csv',index=False)
```
