```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd
```

```python
my_service = Service(executable_path='c:/chrome_driver/chromedriver.exe')
browser = Browser('chrome', service=my_service)
```

```python
url = 'https://shop.gomers.com/shop/?type=beer&order=name+asc'
browser.visit(url)
```

```python
beer_soup = soup(browser.html)
```

```python
for i in beer_soup.find('div',class_='current-store-info').find_all('strong'):
    address = (i.text)
```

```python
contact = beer_soup.find('div',class_='footer-column-content').find('span',class_='contact-us-text').text
```

```python
store_dict = {}
store_dict['Name'] = "Gomer's"
store_dict['Address'] = address
store_dict['Contact'] = contact
store_dict['StoreID'] = 1
gomers_df = pd.DataFrame([store_dict])
```

```python
gomers_df.to_csv('GomersInfo.csv',index = False)
```

```python
beer_list = []
beer_soup = soup(browser.html)
for beer in beer_soup.find_all('div',class_='ch-left-side'):
    beer_dict = {}
    beer_name = beer.find('div',class_='ch-product-name').text
    if ('Boulevard' in beer_name) or ('City Barrel' in beer_name) or ('Kc Bier' in beer_name) or ('Martin City' in beer_name):
        beer_dict['Price'] = beer.find('span').text.replace('$','')
        beer_dict['QTY'] = beer.find('div',class_='ch-product-options-wrapper').find('span').text
        beer_dict['Name'] = beer_name
        beer_list.append(beer_dict)
xpath = '//*[@id="shop-page"]/ch-elements.search-results/div[2]/div[2]/div/div/a'
browser.find_by_xpath(xpath).click()

while True:
    time.sleep(1)
    beer_soup = soup(browser.html)
    for beer in beer_soup.find_all('div',class_='ch-left-side'):
        beer_dict = {}
        beer_name = beer.find('div',class_='ch-product-name').text
        if ('Boulevard' in beer_name) or ('City Barrel' in beer_name) or ('Kc Bier' in beer_name) or ('Martin City' in beer_name):

            beer_dict['Price'] = beer.find('span').text.replace('$','')
            beer_dict['QTY'] = beer.find('div',class_='ch-product-options-wrapper').find('span').text
            beer_dict['Name'] = beer.find('div',class_='ch-product-name').text
            beer_list.append(beer_dict)
    try:
        xpath = '//*[@id="shop-page"]/ch-elements.search-results/div[2]/div[2]/div/div/a[2]'
        browser.find_by_xpath(xpath).click()
    except:
        break
```

```python
df = pd.DataFrame(beer_list)
```
